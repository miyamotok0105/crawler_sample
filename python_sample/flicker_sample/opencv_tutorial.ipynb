```python
#pythonの元から持ってるの
import sys
#opencv
import cv2
#数値計算用
import numpy as np
#matplotlibは表示用のツール
%matplotlib inline
import matplotlib.pyplot as plt

#画像ライブラリの違いによってチャンネルの順番が違うことがある。pillowライブラリ、scipyなど
img = cv2.imread('Grace_Hopper.jpg', cv2.IMREAD_COLOR)

#調査時メソッド
print(img.shape) #配列の形を見る(numpy)
print(img.dtype) #配列の型を見る(numpy)
print(type(img)) #pythonの型


#opencvはgbr
#matplotlibはrgb
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.plot(), plt.imshow(img)
plt.show()
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# print(gray.shape)
# gray = gray.reshape(606, 517,1)
# print(gray.shape)

plt.plot(), plt.imshow(gray)
plt.show()
```

# 顔画像の検出

```python
faceCascade = cv2.CascadeClassifier("haarcascade_frontalface_alt.xml")
faces = faceCascade.detectMultiScale(
    gray,
    scaleFactor=1.1,
    minNeighbors=5,
    minSize=(30, 30),
    flags=cv2.CASCADE_SCALE_IMAGE
)

# Draw a rectangle around the faces
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 2)

    plt.plot(), plt.imshow(img)
plt.show()
```
